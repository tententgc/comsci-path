```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import matplotlib.pyplot as plt
## %matplotlib inline

# Change font size of the chart
# from https://stackoverflow.com/questions/3899980/how-to-change-the-font-size-on-a-matplotlib-plot

SMALL_SIZE  = 12
MEDIUM_SIZE = 16
BIGGER_SIZE = 20

plt.rc('font',  size=SMALL_SIZE)         # controls default text sizes
plt.rc('axes',  titlesize=SMALL_SIZE)    # fontsize of the axes title
plt.rc('axes',  labelsize=MEDIUM_SIZE)   # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
# First perform clustering computer-generated points created with function make_blobs()

from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.9, random_state=0)

plt.scatter(X[:, 0], X[:, 1], s=30)
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=99)  # create the model, try 4 clusters

kmeans.fit(X)     # compute K-mean clustering

y_kmeans = kmeans.predict(X)  # Predict cluster index for each sample
                              # return the .labels_ to y_kmeans
```

```python
print(kmeans.cluster_centers_) # get centroid of each resulted cluster
print(kmeans.labels_)          # see label (cluster number) of each point
```

```python
print(kmeans.n_features_in_)  # get number of features seen during fit()
print(kmeans.n_iter_)
```

```python
# Visualize the clusters
#
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=20, cmap='Set2')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=100, alpha=0.5)
plt.show()
```

## Determine good cluster size using elbow method with silhouette_score()
https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html

The silhouette plot displays a measure of how close each point in one cluster is to points in the neighboring clusters and thus provides a way to assess parameters like number of clusters visually. This measure has a range of [-1, 1].

Silhouette coefficients near +1 indicate that the sample is far away from the neighboring clusters. A value of 0 indicates that the sample is on or very close to the decision boundary between two neighboring clusters and negative values indicate that those samples might have been assigned to the wrong cluster.


```python
from sklearn.cluster import KMeans
from sklearn import metrics

wcss = []
for k in range(2, 11):  # test for 2 to 10 clusters
    kmeans = KMeans(n_clusters=k, random_state=123)
    kmeans.fit(df)
    labels = kmeans.labels_
    wcss.append(kmeans.inertia_)
    print("silhouette_score of k = " + str(k) + " is {:.2f}"
          .format(metrics.silhouette_score(df, labels, metric='euclidean')))

plt.plot(range(2,11), wcss, linewidth=2, color="red", marker ="8")
plt.xticks(np.arange(1,11,1))
plt.xlabel("K value")
plt.ylabel("WCSS")
plt.grid(True)
plt.show()
```

```python
# 1) Create K-Mean clustering
kmeans = KMeans(n_clusters=5, random_state=123)

# 2) Compute K-mean clustering
kmeans.fit(df)
print(kmeans.cluster_centers_)

# 3) Compute cluster index for each point
y_kmeans = kmeans.predict(df)

print(y_kmeans)
```

```python
# Add cluster ID into the data df
df['cluster'] = y_kmeans
print(df)
```

# Cluster analysis with AglomerativeClustering

```python
# First perform clustering computer-generated points created with function make_blobs()

from sklearn.datasets import make_blobs

N = 10

X, y_true = make_blobs(n_samples=N, centers=2, cluster_std=2.5, random_state=123)

# Create Plot
plt.figure(figsize=(8,6))
plt.subplots_adjust(bottom=0.1)
plt.xlabel("X1")
plt.ylabel("X2")

labels = range(1,N+1)

plt.scatter(X[:, 0], X[:, 1], s=30, label='True Position')

for label, x, y in zip(labels, X[:, 0], X[:, 1]):
    plt.annotate(
        label,
        xy=(x,y), xytext=(-3,3),
        textcoords = 'offset points', ha='right', va='bottom')

plt.show()
```

```python
# Import dendrogram and linkage library from scipy
from scipy.cluster.hierarchy import dendrogram, linkage

# Initial figure size
plt.figure(figsize=(8,6))

# Assign labels to represent all data points in X: here N points
labelList = range(1, N+1)

# Create Single-linkage of two points in X
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.linkage.html
linked = linkage(X, 'single')

# Compile dendrogram plot on linked X by using Euclidean distance 
dendrogram(linked, orientation='top', labels=labelList)

plt.title('Single linkage', fontsize=18)
plt.ylabel("Cluster Distance")
plt.xlabel("Data Points")
plt.show()
```

```python
# Initial figure size
plt.figure(figsize=(8,6))

# Create complete-linkage of two points in X
linked = linkage(X, 'complete')

# Compile dendrogram plot on linked X by using Euclidean distance 
dendrogram(linked, orientation='top', labels=labelList)

plt.title('Complete linkage', fontsize=18)
plt.ylabel("Cluster Distance")
plt.xlabel("Data Points")
plt.show()
```

```python
# Initial figure size
plt.figure(figsize=(8,6))

# Create Average-linkage of two points in X
linked = linkage(X, 'average')

# Compile dendrogram plot on linked X by using Euclidean distance 
R = dendrogram(linked, orientation='top', labels=labelList)

plt.title('Average linkage', fontsize=18)
plt.ylabel("Cluster Distance")
plt.xlabel("Data Points")
plt.show()
```

```python
# Initial figure size
plt.figure(figsize=(8,6))

# Create Single-linkage of two points in X
linked = linkage(X, 'ward')

# Compile dendrogram plot on linked X by using Euclidean distance 
dendrogram(linked, orientation='top', labels=labelList)

plt.title('Ward linkage', fontsize=18)
plt.ylabel("Cluster Distance")
plt.xlabel("Data Points")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=3, linkage='ward')

# Call the fit_predict method to returns the cluster ID of each data point
cluster.fit_predict(X)

# Print labels
print(cluster.labels_)

# Plot Scatter
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=cluster.labels_, s=30, cmap='brg')
plt.xlabel("X1")
plt.ylabel("X2")
for label, x, y in zip(labels, X[:, 0], X[:, 1]):
    plt.annotate(
        label,
        xy=(x,y), xytext=(-3,3),
        textcoords = 'offset points', ha='right', va='bottom')
plt.show()
```
